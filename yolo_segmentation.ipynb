```python
import cv2
import numpy as np

def parse_yolo_segmentation_labels(label_path, image_width, image_height):
    """YOLO segmentasyon etiket dosyasÄ±nÄ± ayrÄ±ÅŸtÄ±rÄ±r ve verileri dÃ¶ndÃ¼rÃ¼r."""
    with open(label_path, 'r') as f:
        lines = f.readlines()
    polygons = []
    class_ids = []
    for line in lines:
        data = line.strip().split()
        class_id = int(data[0])
        points = [(float(x) * image_width if i % 2 == 0 else float(x) * image_height)
                  for i, x in enumerate(data[1:])]
        polygons.append(np.array(points).reshape(-1, 2).astype(int))
        class_ids.append(class_id)
    return polygons, class_ids

label_path = "/content/data/test/labels/train_49_png.rf.56c1159720608fdc7726fe61b4ceeba6.txt"

image_path = "/content/data/test/images/train_49_png.rf.56c1159720608fdc7726fe61b4ceeba6.jpg"
image = cv2.imread(image_path)

image_width = image.shape[1]
image_height = image.shape[0]

polygons, class_ids = parse_yolo_segmentation_labels(label_path, image_width, image_height)

class_colors = {
    0 : (0, 255, 0), # yeÅŸil
    1 : (255, 0, 0), # kÄ±rmÄ±zÄ±
    2 : (0, 0, 255), # mavi
    3 : (255, 255, 0), # sarÄ±
}

mask = np.zeros_like(image)
for polygon, class_id in zip(polygons, class_ids):
    cv2.fillPoly(mask, [polygon], class_colors[class_id])

result = cv2.addWeighted(image, 0.7, mask, 0.3, 0)

cv2_imshow(result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
